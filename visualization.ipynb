# Visualisasi Data Dasar
**Menggunakan Matplotlib dan Seaborn Dasar**

```python
# mengakses library matplotlib dan seaborn untuk visualisasi data
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading Dataset

```python
# mendownload dataset 'tips' dari seaborn yang sudah tersedia
tips = sns.load_dataset('tips')
tips.head()
# mengecek isi dari dataset dengan memperlihatkan 5 baris data teratas
```

# 1. Histogram

```python
tips.info()
# memperlihatkan informasi ringkas tentang dataset
# memperlihatkan nama kolom, tipe data, dan jumlah nilai non-null pada setiap kolom
```

```python
tips.describe(include='all').transpose()
# memberikan ringkasan statistik deskriptif dari dataset
# include='all' berarti semua tipe data akan disertakan dalam ringkasan statistik
# transpose() digunakan untuk memudahkan pembacaan hasil ringkasan statistik dengan menukar baris dan kolom
```

```python
# membuat histogram, yaitu representasi grafis berbentuk batang yang menunjukkan distribusi frekuensi data
sns.histplot(data=tips, x='sex', shrink=.8)
plt.title('Gender', size=16)
plt.show()
# menggunakan fungsi histplot dari seaborn untuk membuat histogram
# data=tips menunjukkan bahwa data yang digunakan berasal dari dataframe tips
# x='sex' menunjukkan bahwa kolom sex digunakan untuk sumbu x pada histogram
# shrink=.8 digunakan untuk mengatur jarak antar batang histogram
# plt.title digunakan untuk memberikan judul pada grafik dan menentukan ukuran font
# jika mennggunakan library matplotlib, maka perlu menambahkan plt.show() untuk menampilkan grafik
```

```python
sns.histplot(data=tips, x='smoker', shrink=.8)
plt.title('Smoker', size=16)
plt.show()
# membuat histogram untuk kolom 'smoker' dengan cara yang sama seperti sebelumnya
```

```python
# membuat histogram menggunakan pandas bawaan dan matplotlib
tips['total_bill'].hist(color='green', alpha=0.4)
plt.title('Bill Distribution', size=16)
plt.show()
# menggunakan fungsi hist dari pandas untuk membuat histogram
# bisa langsung mengakses kolom dari dataframe tips dengan cara tips['total_bill']
# color='green' digunakan untuk mengatur warna batang histogram
# alpha=0.4 digunakan untuk mengatur tingkat transparansi warna batang histogram
```

```python
# membuat histogram dengan memisahkan data berdasarkan kolom
plt.figure(figsize=(10, 6))
sns.histplot(data=tips, x='total_bill', hue='sex')
plt.title('Bill Distribution Based on Sex', size=14)
plt.show()
# menggunakan parameter hue untuk memisahkan data berdasarkan kolom
# fungsi figure dar matplotlib digunakan untuk mengatur ukuran grafik
```

# 2. Boxplot

```python
# membuat boxplot, yaitu representasi grafis berbentuk kotak yang menunjukkan distribusi data berdasarkan kuartil
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Boxplot of Total Bill by Day', size=16)
plt.show()
# menentukan sumbu x dan y dengan kolom 'day' dan 'total_bill' dari dataframe tips
# pada grafik bloxplot:
# garis horizontal di dalam kotak menunjukkan median
# batas atas dan bawah kotak menunjukkan kuartil pertama (Q1) dan kuartil ketiga (Q3)
# garis vertikal (whiskers) menunjukkan rentang data atau nilai minimum dan maksimum dalam batas tertentu
# titik-titik di luar whiskers menunjukkan outlier atau nilai yang jauh dari distribusi data utama
```

```python
sns.boxplot(x='time', y='tip', data=tips, order=['Dinner', 'Lunch'])
plt.title('Boxplot of Tip by Time', size=16)
plt.show()
# menggunakan parameter order untuk mengatur urutan kategori pada sumbu x
```

```python
sns.boxplot(x='day', y='tip', data=tips)
plt.title('Boxplot of Tip by Day', size=16)
plt.show()
# ini menunjukkan tips yang paling tinggi diberikan pada hari Sabtu (Sat) dibandingkan hari lainnya
```

# 3. Lineplot

```python
# mendownload dataset 'flights' dari seaborn yang sudah tersedia
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.info()
# memperlihatkan informasi ringkas tentang dataset flights
# memperlihatkan nama kolom, tipe data, dan jumlah nilai non-null pada setiap kolom
```

```python
flights.describe(include='all')
# memberikan ringkasan statistik deskriptif dari dataset flights
# untuk semua tipe data
```

```python
# membuat lineplot, yaitu representasi grafis berbentuk garis yang menunjukkan tren data dari waktu ke waktu
sns.lineplot(data=flights, x='year', y='passengers')
plt.title('Number of Passengers Over Years', size=16)
plt.show()
# menggunakan fungsi lineplot dari seaborn untuk membuat lineplot
# menggunakan data dari dataframe flights
# memasukkan kolom 'year' sebagai sumbu x dan kolom 'passengers' sebagai sumbu y
```

```python
sns.lineplot(data=flights, x='month', y='passengers')
plt.title('Number of Passengers Over Months', size=16)
plt.show()
# membuat lineplot untuk melihat tren jumlah penumpang berdasarkan bulan
```

```python
january_flights = flights.query("month == 'Jan'")
january_flights
# menggunakan metode query dari pandas untuk memfilter data
# memfilter data untuk hanya menampilkan baris dengan bulan Januari (Jan) saja
```

```python
sns.lineplot(data=january_flights, x='year', y='passengers')
plt.title('Number of Passengers in January Over Years', size=16)
plt.show()
# membuat lineplot untuk melihat tren jumlah penumpang pada bulan Januari dari tahun ke tahun
```

```python
sns.lineplot(data=flights, x='year', y='passengers', hue='month')
plt.title('Number of Passengers Over Years Based on Month', size=14)
plt.show()
# membuat lineplot dengan memisahkan data berdasarkan bulan dengan parameter hue
# pemisahan terlihat dari warna garis yang berbeda untuk setiap bulan
```

# 4. Scatter Plot

```python
# membuat scatterplot, yaitu representasi grafis berbentuk titik yang menunjukkan hubungan antara dua atau lebih variabel
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time')
plt.title('Total Bill vs Tip', size=15)
plt.show()
# menggunakan fungsi scatterplot dari seaborn untuk membuat scatterplot
# memasukkan kolom 'total_bill' sebagai sumbu x dan kolom 'tip' sebagai sumbu y
# menggunakan parameter hue untuk memisahkan data berdasarkan kolom 'day'
# menggunakan parameter style untuk memisahkan data berdasarkan kolom 'time' dengan bentuk titik yang berbeda
```
